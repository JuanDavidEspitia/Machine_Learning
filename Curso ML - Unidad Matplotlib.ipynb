# MatplotLib

- Es una biblioteca para la generacion de graficos apartir de datos contenidos en listas o arrays en el lenguaje de programacion Python
- Matplotlib te da control sobre cada aspecto del grafico
- Los graficos que posee la libreria son:
https://matplotlib.org/3.2.1/gallery/index.html

```python
import pandas as pd
import numpy as np
import seaborn as sns                      
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(2,9,10) # Una lsista de valores que van desde el 2 hasta 9 y toma 10 elementos
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y) # Grafica sencilla en el plano cartesiano
```

```python
plt.plot(x,y,'red') # Cambiamo el color de la grafica
```

```python
plt.plot(x,y,'green') # Cambiamo el color de la grafica
```

```python
plt.plot(x,y,'green') # Podemos ponerle un titulo
plt.title('Grafico ejemplo')
```

```python
plt.plot(x,y,'green') 
plt.title('Grafico ejemplo') # Podemos ponerle un titulo
# Le ponemos etiquetas en x e y
plt.xlabel('Eje x')
plt.ylabel('Eje y')
```

```python
# Podemos aadir dos graficos pequeos 
plt.subplot(1,2,1)
plt.plot(x,y,'red')

plt.subplot(1,2,2)
plt.plot(y,x,'green')
```

```python
# Podemos aadir dos graficos pequeos  
plt.subplot(2,1,1)
plt.plot(x,y,'red')

plt.subplot(2,1,2)
plt.plot(y,x,'green')
```

Programacion orientada a objetos

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.7,0.7])
grafico.plot(x,y,'green')
```

```python
figura = plt.figure()
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'green')
```
